```python
import pandas as pd 
import numpy as np


weather_data = {
    "city": [
        "Kathmandu", "Pokhara", "Delhi", "Tokyo", "London",
        "New York", "Sydney", "Dubai"
    ],
    
    "date": [
        "2026-03-01",
        "2026-03-02",
        "2026-03-04",   # 03 missing
        "2026-03-05",
        "2026-03-07",   # 06 missing
        "2026-03-08",
        "2026-03-10",   # 09 missing
        "2026-03-12"    # 11 missing
    ],
    
    "temperature_c": [
        18, 20, np.nan, 16, 12, 10, 22, np.nan
    ],
    
    "humidity_percent": [
        60, np.nan, 50, 55, 70, "Not Available", 58, 40
    ],
    
    "wind_speed_kmh": [
        10, 8, 12, np.nan, 20, 18, "Not Available", 9
    ],
    
    "weather_condition": [
        "Cloudy", "Rainy", "Sunny", np.nan,
        "Foggy", "Snowy", "Sunny", "Hot"
    ]
}
```

```python
df  = pd.DataFrame(weather_data)
df["date"] = pd.to_datetime(df["date"])
df.replace("Not Available", np.nan, inplace=True)
df.set_index("date", inplace=True)
df
```

```python
dt = pd.date_range(start = "2026-03-01", end = "2026-03-12" , freq="D")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df


## now after this we can apply the interpolation method to fill the null values in the date column as well as other columns
```

```python
new_df = df.fillna({
    "temperature_c": df["temperature_c"].mean(),
    "humidity_percent": df["humidity_percent"].interpolate()[0],
    "wind_speed_kmh": df["wind_speed_kmh"].median(),
    "weather_condition": "Unknown"
})
new_df
```
